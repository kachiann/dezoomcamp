```python
import os
import subprocess

yellow_path = "yellow_tripdata_2025-11.parquet"
zones_path = "taxi_zone_lookup.csv"

if not os.path.exists(yellow_path):
    print("Downloading yellow_tripdata_2025-11.parquet...")
    subprocess.run([
        "wget",
        "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-11.parquet"
    ], check=True)
else:
    print("Yellow data already present.")

if not os.path.exists(zones_path):
    print("Downloading taxi_zone_lookup.csv...")
    subprocess.run([
        "wget",
        "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
    ], check=True)
else:
    print("Zone lookup already present.")
```

### Q2: Repartition the Dataframe to 4 partitions and save it to parquet.

```python
# Read original data
df = spark.read.parquet(yellow_path)
print("Original rows:", df.count())

# Repartition and write
output_dir_q2 = "yellow_2025_11_repart_4"
df_repart = df.repartition(4)
df_repart.write.mode("overwrite").parquet(output_dir_q2)

print("Written to:", output_dir_q2)
```
